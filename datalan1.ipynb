```python
from PIL import Image, ImageDraw
import matplotlib.pyplot as plt

image_path = "/content/runs/detect/predict/yolov7test4_mp4-187_jpg.rf.4fa37c86e44c8a96445dc47020f4fa23.jpg"
label_path = "/content/runs/detect/predict/labels/yolov7test4_mp4-187_jpg.rf.4fa37c86e44c8a96445dc47020f4fa23.txt"

image = Image.open(image_path)
draw = ImageDraw.Draw(image)
width, height = image.size
```

```python

bounding_boxes = []
with open(label_path, "r") as f:
    for line in f:
        parts = line.strip().split()
        class_id, x_center, y_center, w, h = map(float, parts)
        x1 = int((x_center - w / 2) * width)
        y1 = int((y_center - h / 2) * height)
        x2 = int((x_center + w / 2) * width)
        y2 = int((y_center + h / 2) * height)
        bounding_boxes.append(((x1, y1, x2, y2), class_id))


feeding_zone = (0, 0, width, int(height * 0.35))
drinking_zone = (0, int(height * 0.35), width, int(height * 0.55))
resting_zone = (0, int(height * 0.55), width, height)


draw.rectangle(feeding_zone, outline="yellow", width=3)
draw.rectangle(drinking_zone, outline="red", width=3)
draw.rectangle(resting_zone, outline="green", width=3)


feeding_count = 0
drinking_count = 0
resting_count = 0

for (x1, y1, x2, y2), class_id in bounding_boxes:
    cx = (x1 + x2) // 2
    cy = (y1 + y2) // 2
    if feeding_zone[1] <= cy < feeding_zone[3]:
        feeding_count += 1
    elif drinking_zone[1] <= cy < drinking_zone[3]:
        drinking_count += 1
    elif resting_zone[1] <= cy <= resting_zone[3]:
        resting_count += 1


draw.text((10, 10), f"Feeding zone: {feeding_count}", fill="yellow")
draw.text((10, 30), f"Drinking zone: {drinking_count}", fill="red")
draw.text((10, 50), f"Resting zone: {resting_count}", fill="green")


plt.figure(figsize=(12, 8))
plt.imshow(image)
plt.axis('off')
plt.title("Phân bố gà theo vùng hoạt động")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Dữ liệu
zones = ['Feeding', 'Drinking', 'Resting']
counts = [feeding_count, drinking_count, resting_count]

# Vẽ biểu đồ
plt.figure(figsize=(6, 4))
plt.bar(zones, counts, color=['yellow', 'red', 'green'])
plt.title('Số lượng gà theo vùng hoạt động')
plt.ylabel('Số lượng gà')
plt.show()
```

```python
total = feeding_count + drinking_count + resting_count
labels = ['Feeding', 'Drinking', 'Resting']
sizes = [feeding_count/total, drinking_count/total, resting_count/total]
colors = ['yellow', 'red', 'green']

plt.figure(figsize=(5, 5))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%')
plt.title('Tỉ lệ gà trong từng vùng hoạt động')
plt.show()
```
